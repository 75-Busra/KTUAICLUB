```python
class calcul(object):
    def __init__(self,list1):
        self.list1 = list1
        self.dict1 = {y:0 for y in self.list1}

    def calculate(self):
        for x in self.list1:
            self.dict1[x] +=1
        return self.dict1

a = calcul(["python","law","artificial","intelligence","software","python","law","artificial","intelligence","software","artificial","intelligence","software","ktuaiclub"]).calculate()
print(a)
```
